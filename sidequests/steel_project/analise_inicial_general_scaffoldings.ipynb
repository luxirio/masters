```python
moleculas = pd.read_csv("VEHICLe.csv")

print(f'O banco de dados possui {moleculas.shape[0]} moleculas')
moleculas.head()
```

O banco de dados fornecido pelo professor Steel possui mais de 24000 estruturas. A matriz fornecida possui algumas variaveis que ainda não compreendemos, i.e (Pgood? Tauto? Training_dataset_hits?). Para nossa análise inicial, usaremos apenas o índice (e Regid) bem como os smiles.

```python
moleculas = moleculas[["Regid","Smiles"]] #selecionamos apenas as duas variaveis

moleculas.dropna(axis=0, how = "any", inplace = True) #dropamos as linhas sem estruturas (aparentemente nenhuma)

moleculas.drop_duplicates("Regid", keep="first", inplace = True) #dropamos os smiles repetidos

moleculas.reset_index(drop = True, inplace = True) #resetamos o índice (isso é realmente necessario @arturcgs?)

print(f'O banco de dados possui {moleculas.shape[0]} moleculas')
moleculas.head()
```

```python
#vamos visualizar

PandasTools.AddMoleculeColumnToFrame(moleculas, "Smiles")
moleculas
#aparentemente ja sao fragmentos!
```

```python
#calculando os descritores de cada um dos "fragmentos"

moleculas["mw"] = moleculas["ROMol"].apply(Descriptors.ExactMolWt) #peso molecular
moleculas["n_hba"] = moleculas["ROMol"].apply(Descriptors.NumHAcceptors)
moleculas["n_hbd"] = moleculas["ROMol"].apply(Descriptors.NumHDonors)
moleculas["logp"] = moleculas["ROMol"].apply(Descriptors.MolLogP)
```

---
<h2>Análise exploratória dos fragmentos!</h2>

Aqui começaremos a fazer análises exploratórias dos fragmentos disponíveis.
Primeiramente, calculamos alguns descritores fundamentais como peso molecular, número de aceptores/doadores de hidrogênio e logP calculado


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")
sns.set_context("talk")

#plot mw
fig, axs = plt.subplots(2,2, figsize = (10,10))
sns.histplot(data = moleculas, x = "mw", kde = False, color = "skyblue", ax = axs[0,0], 
             bins = 30).set(title='Molecular Weight (MW) distribution', xlabel = "")
axs[0,0].axvline(moleculas.mw.mean(), color = "red") 
axs[0,0].axvline(300, color = "blue", linestyle = "--") #max frag based

#plot LogP
sns.histplot(data = moleculas, x = "logp", kde = False, color = "olive", ax = axs[0,1], 
             bins = 30).set(title='LogP distribution' , xlabel = "", ylabel= "")
axs[0,1].axvline(moleculas.logp.mean(), color = "red")
axs[0,1].axvline(3, color = "blue", linestyle = "--") #max frag based

#plot numero de doadores de h
sns.histplot(data = moleculas, x = "n_hbd", kde = False, color = "gold", ax = axs[1,0],
             binwidth = 1).set(title = "Number of Hydrogen bond \ndonor distribution", xlabel = "")
axs[1,0].axvline(moleculas.n_hbd.mean(), color = "red")
axs[1,0].axvline(3, color = "blue", linestyle = "--") #max frag based


#plot numero de aceptores de h
sns.histplot(data = moleculas, x = "n_hba", kde = False, color = "teal", ax = axs[1,1], 
             binwidth = 1).set(title = "Number of Hydrogen bond \nacceptor distribution", xlabel = "", ylabel= "")
axs[1,1].axvline(moleculas.n_hba.mean(), color = 'red')
axs[1,1].axvline(3, color = "blue", linestyle = "--") #max frag based

fig.tight_layout()
```

Podemos observar que a distribuicao de peso molecular, logp, numero de aceptores de hidrogenio possuem distribuicao aproximadamente normal!

```python
print(f'media mw = {moleculas.mw.mean()}')
print(f'media logp = {moleculas.logp.mean()}')
print(f'media hbd = {moleculas.n_hbd.mean()}')
print(f'media hba = {moleculas.n_hba.mean()}')
```

```python
sns.set(style = "white")
sns.set_context("talk")

mw_logp_plt = sns.jointplot(x = "mw", y = "logp", data = moleculas, 
              alpha = 0.5, height = 12, s = 30).set_axis_labels("Molecular Weight", "(Calculated) LogP")

mw_logp_plt.figure.tight_layout()
```
